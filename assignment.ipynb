```python
import tensorflow
```

```python
from tensorflow.keras.datasets import boston_housing
```

```python
(feat,prices),_=boston_housing.load_data(test_split=0)
```

```python
print('No of examples',feat.shape)
```

```python
model=tensorflow.keras.models.Sequential()
```

```python
model.add(tensorflow.keras.layers.BatchNormalization(input_shape=(13,)))
```

```python
model.add(tensorflow.keras.layers.Dense(1))
```

```python
print(model.summary())
```

```python
model.compile(optimizer='sgd',loss='mse')
```

```python
model.fit(feat,prices,epochs=100,validation_split=0.3,verbose=2)
```
